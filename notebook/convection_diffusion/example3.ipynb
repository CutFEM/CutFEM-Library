### Problem

This is the N-sphere problem from Lehrenfeld, but where the solution is given on the form $u(\pi\frac{r^2}{r_0^2})\sin(\pi t)$ instead of $u(\pi \frac{r}{r_0})\sin(\pi t)$ as in the original example. The hypthesis here is that we need the form of the level-set function (Which is on the former form) to match the form in the solution.

We therefore run with the same parameters as the N-sphere case, i.e. 
* h = [0.9, 0.45, 0.225, 0.1125, 0.05625, 0.028125, 0.0140625],
* dT = [0.25, 0.125, 0.0625, 0.03125, 0.015625, 0.0078125, 0.00390625]

* T = 0.5
* tau = 0.05
* D = 1

This of course leads to a different RHS function, but the Neumann boundary condition still remains zero.
